```python
from core.data.loaders import CompetitionDataLoader
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv("competition_data/digit-recognizer/train.csv")
```

```python
# drop columns that have a 0.0 mean
df = df.loc[:, (df != 0).any(axis=0)]
df.tail(10)
```

```python
path = Path("competition_data/titanic")
competition_data_loader = CompetitionDataLoader(path)
feature_types = competition_data_loader.analyze_feature_types(df)
# print feature types where value is binary
for col, ftype in feature_types.items():
    if ftype == "binary":
        print(f"{col}: {ftype}")
        # Unique values in the column
        print(df[col].unique())
```
